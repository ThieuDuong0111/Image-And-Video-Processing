<h3>Draw Shapes on Images</h3>

We need to understand the required functions in OpenCV-Python, which help us to draw the shapes on images.

<table>
        <thead>
            <tr>
                <th>Function</th>
                <th>Description</th>
                <th>Command</th>
            </tr>
        </thead>
        <tbody>
            <tr>
                <td><code>cv2.line()</code></td>
                <td>Draws a line segment connecting two points.</td>
                <td><code>cv2.line(img, pt1, pt2, color, thickness)</code></td>
            </tr>
            <tr>
                <td><code>cv2.circle()</code></td>
                <td>Draws a circle of given radius at given point as center.</td>
                <td><code>cv2.circle(img, center, radius, color, thickness)</code></td>
            </tr>
            <tr>
                <td><code>cv2.rectangle()</code></td>
                <td>Draws a rectangle with given points as top-left and bottom-right.</td>
                <td><code>cv2.rectangle(img, pt1, pt2, color, thickness)</code></td>
            </tr>
            <tr>
                <td><code>cv2.ellipse()</code></td>
                <td>Draws a simple or thick elliptic arc or fills an ellipse sector.</td>
                <td><code>cv2.ellipse(img, center, axes, angle, startAngle, endAngle, color, thickness)</code></td>
            </tr>
        </tbody>
    </table>

```python
import numpy as np
import cv2

img = cv2.imread('../opencv_logo.png',1)

cv2.line(img,(20,400),(400,20),(255,255,255),3)
cv2.rectangle(img,(200,100),(400,400),(0,255,0),5)
cv2.circle(img,(80,80), 55, (255,255,0), -1)
cv2.ellipse(img, (300,425), (80, 20), 5, 0, 360, (0,0,255), -1)
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('../opencv_logo.png',1)
txt="OPENCV Logo"
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img,txt,(10,100), font, 2,(255,255,255),2,cv2.LINE_AA)

cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
